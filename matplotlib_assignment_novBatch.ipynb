We have the min and max temperatures in a city In India for each months of the year.
We would like to find a function to describe this and show it graphically, the dataset
given below.

Task:
    
1.fitting it to the periodic function

2.plot the fit

Data
Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25

Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

Max = [39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25]
Min = [21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18]
X = np.arange(12)
tempdata = pd.DataFrame(list(zip(Max, Min)), columns=['Max', 'Min'])
plt.figure(figsize=(6,7))
plt.plot(X, Max,'ro')
plt.plot(X, Min, 'bo')
plt.xlabel('Month')
plt.ylabel('Temperature ($^\circ$C)')
```

Fitting it to the periodic function

```python
from scipy import optimize

def yearly_temps(times, avg, ampl, time_offset):
    return (avg
            + ampl * np.cos((times + time_offset) * 2 * np.pi / times.max()))

res_max, cov_max = optimize.curve_fit(yearly_temps, X, Max, [20, 10, 0])
res_min, cov_min = optimize.curve_fit(yearly_temps, X, Min, [-40, 20, 0])
```

plot the fit

```python
days = np.linspace(0, 12, num=1005)

plt.figure(figsize=(8,6))
plt.plot(X, Max, 'ro')
plt.plot(days, yearly_temps(days, *res_max), 'r-')
plt.plot(X, Min, 'bo')
plt.plot(days, yearly_temps(days, *res_min), 'b-')
plt.xlabel('Month')
plt.ylabel('Temperature ($^\circ$C)')
```
